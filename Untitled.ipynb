```python
import SimpleITK as sitk
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
#file path to volume
fn_hist = "../data/read/25/hist_1_affine_rgb.nii" 
```

```python
#histology volume
im_hist= sitk.ReadImage(fn_hist,sitk.sitkVectorUInt8)
print(im_hist.GetSize(), im_hist.GetSpacing(), im_hist.GetNumberOfComponentsPerPixel())
```

```python
lenth = im_hist.GetSize()[2] # for all slices
#print(lenth)
for i in range(0,lenth):
    slice_idx = i
    #slice_idx = 10
    slice_hist = im_hist[:,:,slice_idx]
    # output directoru
    fn_slice = "../data/write/25/im" + str(slice_idx) + ".png" # changes file name and all
    arr = sitk.GetArrayFromImage(im_hist)
    slice_hist_arr = arr[slice_idx,:,:,:]

    # can I add the downsampling logic here
    slice_hist_arr = cv2.GaussianBlur(slice_hist_arr, (0, 0), 1,1)
    slice_hist_arr_1 = cv2.resize(slice_hist_arr, (256,165), interpolation=cv2.INTER_CUBIC)
    plt.imshow(slice_hist_arr_1); plt.show()
    slice_hist_arr_1.shape
    cv2.imwrite(fn_slice,cv2.cvtColor(slice_hist_arr_1, cv2.COLOR_RGB2BGR))
    
```
